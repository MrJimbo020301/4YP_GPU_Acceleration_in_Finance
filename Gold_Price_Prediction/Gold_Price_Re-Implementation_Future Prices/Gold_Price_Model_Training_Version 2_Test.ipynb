```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import os
import random
import warnings

# ------------------------------------------------------------------------
# 1. SET RANDOM SEEDS
# ------------------------------------------------------------------------
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(0)

# ------------------------------------------------------------------------
# 2. LOAD AND PREPARE DATA
# ------------------------------------------------------------------------
df = pd.read_csv('Gold Futures Historical Data_Original.csv')
df.drop(['Vol.', 'Change %'], axis=1, inplace=True)

df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)

num_cols = df.columns.drop('Date')
df[num_cols] = df[num_cols].replace({',': ''}, regex=True).astype('float64')

# Check for duplicates/nulls
assert df.duplicated().sum() == 0, "Duplicate rows found!"
assert df.isnull().sum().sum() == 0, "Missing values found!"

# We will treat all four columns as features to be predicted:
# 0: Price, 1: Open, 2: High, 3: Low
features = ['Price', 'Open', 'High', 'Low']

# ------------------------------------------------------------------------
# 3. SPLIT BY DATE (TRAIN / VALIDATION / TEST) BEFORE SCALING
# ------------------------------------------------------------------------
train_cutoff = pd.to_datetime('2023-10-24')
val_cutoff   = pd.to_datetime('2024-11-28')

df_train = df[df['Date'] < train_cutoff]                # up to 2018-10-23
df_val   = df[(df['Date'] >= train_cutoff) & (df['Date'] < val_cutoff)]  # 2018-10-24 to 2018-11-27
df_test  = df[df['Date'] >= val_cutoff]                 # from 2018-11-28 onward

# ------------------------------------------------------------------------
# 4. FIT SCALER ONLY ON TRAINING DATA (ALL 4 FEATURES)
# ------------------------------------------------------------------------
scaler = MinMaxScaler()
scaler.fit(df_train[features])  # scale params from train only

# Apply transform to each split
df_train_scaled = df_train.copy()
df_val_scaled   = df_val.copy()
df_test_scaled  = df_test.copy()

df_train_scaled[features] = scaler.transform(df_train[features])
df_val_scaled[features]   = scaler.transform(df_val[features])
df_test_scaled[features]  = scaler.transform(df_test[features])

# Also create a combined scaled DataFrame for convenience
df_scaled = pd.concat([df_train_scaled, df_val_scaled, df_test_scaled], ignore_index=True)

# ------------------------------------------------------------------------
# 5. BUILD (X, y) FOR TRAIN & VAL (SINGLE-STEP, MULTI-OUTPUT)
#    Each target y is [Price, Open, High, Low] at day t.
#    X is the 30 previous days of these 4 features.
# ------------------------------------------------------------------------
window_width = 30

def make_sequences(df_subset_scaled, window=30):
    arr = df_subset_scaled[features].values  # shape [N, 4]
    dates_arr = df_subset_scaled['Date'].values  # shape [N, ]

    X_list, y_list, y_dates_list = [], [], []
    for i in range(window, len(arr)):
        X_window = arr[i - window : i]   # shape [30, 4]
        y_target = arr[i]               # shape [4, ] (all features)
        X_list.append(X_window)
        y_list.append(y_target)
        y_dates_list.append(dates_arr[i])
    return np.array(X_list), np.array(y_list), np.array(y_dates_list)

X_train_all, y_train_all, train_dates_all = make_sequences(df_train_scaled, window_width)
X_val_all,   y_val_all,   val_dates_all   = make_sequences(df_val_scaled,   window_width)

# Convert to PyTorch Tensors
X_train_tensor = torch.from_numpy(X_train_all).float()          # [N_train, 30, 4]
y_train_tensor = torch.from_numpy(y_train_all).float()          # [N_train, 4]

X_val_tensor   = torch.from_numpy(X_val_all).float()            # [N_val, 30, 4]
y_val_tensor   = torch.from_numpy(y_val_all).float()            # [N_val, 4]

# Build DataLoaders
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset   = TensorDataset(X_val_tensor,   y_val_tensor)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
val_loader   = DataLoader(val_dataset,   batch_size=batch_size, shuffle=False)

# ------------------------------------------------------------------------
# 6. DEFINE MULTI-OUTPUT MODEL CLASS
#    We'll produce 4 outputs (Price, Open, High, Low) at the final step.
# ------------------------------------------------------------------------
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class BaseModel(nn.Module):
    def __init__(self, model_type='LSTM'):
        super().__init__()
        self.model_type = model_type
        
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=2,
                               batch_first=True, dropout=0.1)
            self.fc = nn.Linear(128, 4)  # 4 outputs
        elif model_type == 'GRU':
            self.rnn = nn.GRU(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, dropout=0.1)
            self.fc = nn.Linear(128, 4)
        elif model_type == 'RNN':
            self.rnn = nn.RNN(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, nonlinearity='relu', dropout=0.1)
            self.fc = nn.Linear(128, 4)
        elif model_type == 'CNN':
            self.conv1 = nn.Conv1d(in_channels=4, out_channels=64, kernel_size=3)
            self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3)
            # Each Conv1d reduces sequence length by (kernel_size - 1)
            conv_output_length = window_width - 2*(3 - 1)  # two conv layers
            self.fc = nn.Linear(128 * conv_output_length, 4)
        elif model_type == 'EnhancedLSTM':
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=3,
                               batch_first=True, dropout=0.2)
            self.bn = nn.BatchNorm1d(128)
            self.dropout = nn.Dropout(0.2)
            self.fc = nn.Linear(128, 4)
        else:
            raise ValueError("Invalid model_type.")
        
    def forward(self, x):
        """
        x shape: [batch_size, seq_len=30, features=4]
        Output: [batch_size, 4] for (Price, Open, High, Low).
        """
        if self.model_type == 'CNN':
            # (batch, seq_len, channels) -> (batch, channels, seq_len)
            x = x.permute(0, 2, 1)
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(x.size(0), -1)  # flatten
            out = self.fc(x)
        
        elif self.model_type == 'EnhancedLSTM':
            out, _ = self.rnn(x)       # out: [batch, seq_len, hidden]
            out = out[:, -1, :]       # take final time step
            out = self.bn(out)
            out = self.dropout(out)
            out = self.fc(out)        # [batch, 4]
        
        else:
            out, _ = self.rnn(x)      # [batch, seq_len, hidden]
            out = out[:, -1, :]       # last time step
            out = self.fc(out)        # [batch, 4]

        return out


class EarlyStopping:
    def __init__(self, patience=150, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0


# ------------------------------------------------------------------------
# 7. HELPER: INVERSE TRANSFORM 4 FEATURES
#    We'll feed [price_scaled, open_scaled, high_scaled, low_scaled] to invert them all.
# ------------------------------------------------------------------------
def inverse_transform_4cols(y_scaled_4):
    """
    y_scaled_4: (N,4) array of scaled [Price, Open, High, Low].
    Return the unscaled (N,4).
    """
    y_scaled_4 = np.array(y_scaled_4, dtype=float)
    # Use the same MinMaxScaler (trained on the train set) to invert.
    return scaler.inverse_transform(y_scaled_4)

# For convenience, if we only want the Price from a 4-col array:
def extract_price_col(y_array_4):
    """Given shape [N, 4], extract the Price column (index 0)."""
    return y_array_4[:, 0]

# ------------------------------------------------------------------------
# 8. TRAIN FUNCTION (ON TRAIN + VAL) - MULTI-OUTPUT
# ------------------------------------------------------------------------
def train_model(model_type='LSTM'):
    model = BaseModel(model_type=model_type).to(device)
    
    # We'll use Huber or MSE for 4-output regression
    criterion = nn.MSELoss()  # or nn.HuberLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.0005)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=150)
    early_stopping = EarlyStopping(patience=150, min_delta=0)
    
    num_epochs = 800
    best_val_loss = float('inf')
    best_model_path = f'best_model_{model_type}_o3.pt'
    
    # If there's a checkpoint, load it
    if os.path.exists(best_model_path):
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=FutureWarning)
            model.load_state_dict(torch.load(best_model_path, map_location=device))
        print(f"Loaded existing model for {model_type} from {best_model_path}")

    for epoch in range(num_epochs):
        # TRAIN
        model.train()
        train_losses = []
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            out = model(X_batch)        # out shape [batch,4]
            loss = criterion(out, y_batch)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
        avg_train_loss = np.mean(train_losses)

        # VALIDATION
        model.eval()
        val_losses = []
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                val_out = model(X_batch)
                val_loss = criterion(val_out, y_batch)
                val_losses.append(val_loss.item())
        avg_val_loss = np.mean(val_losses)

        scheduler.step(avg_val_loss)
        early_stopping(avg_val_loss)

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), best_model_path)

        if early_stopping.early_stop:
            print(f"[{model_type}] Early stopping at epoch {epoch + 1}")
            break

        if (epoch + 1) % 100 == 0:
            print(f"[{model_type}] Epoch {epoch+1}/{num_epochs}, "
                  f"Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")

    # Load best weights
    model.load_state_dict(torch.load(best_model_path, map_location=device))
    return model


# ------------------------------------------------------------------------
# 9. WALK-FORWARD PREDICTION (MULTI-OUTPUT) FOR ANY DATE RANGE
#    Overwrites [Price,Open,High,Low] in the 'df_work' once we pass
#    the 'start_date'. That way, future days see predictions only.
# ------------------------------------------------------------------------
def walk_forward_predict(model, df_all_scaled, start_date, end_date, window=30):
    """
    model: trained (frozen) multi-output model
    df_all_scaled: entire scaled DataFrame (train+val+test)
    start_date, end_date: range for walk-forward
    window: 30 days
    Returns a DataFrame with columns:
      Date, [Pred_Price, Pred_Open, Pred_High, Pred_Low] (scaled),
      plus the inverse-transformed columns as well.
    """
    df_work = df_all_scaled.copy()
    df_work.sort_values('Date', inplace=True)

    # Make sure columns are in the correct order
    # We'll store predictions in the row's scaled features once a date is >= start_date
    predictions_list = []

    dates_in_range = df_work[(df_work['Date'] >= start_date) & 
                             (df_work['Date'] <= end_date)]['Date'].unique()
    dates_in_range = np.sort(dates_in_range)

    model.eval()
    with torch.no_grad():
        for d in dates_in_range:
            # Find the index of this date in df_work
            idx_current = df_work.index[df_work['Date'] == d]
            if len(idx_current) == 0:
                continue
            idx_current = idx_current[0]

            # We need 30 prior days to form X
            start_idx = idx_current - window
            if start_idx < 0:
                continue
            
            # Build input window from the *already updated* df_work
            window_data = df_work.iloc[start_idx:idx_current][features].values  # shape [30,4]
            # shape -> [1,30,4] for the model
            X_input = torch.from_numpy(window_data).float().unsqueeze(0).to(device)
            
            pred_4 = model(X_input).cpu().numpy()[0]  # shape [4,], scaled
            # Overwrite the scaled [Price,Open,High,Low] for date d in df_work
            df_work.loc[idx_current, features] = pred_4

            predictions_list.append((d, pred_4[0], pred_4[1], pred_4[2], pred_4[3]))

    # Build DataFrame
    pred_df = pd.DataFrame(predictions_list, columns=[
        'Date', 'Pred_Price_scaled', 'Pred_Open_scaled', 'Pred_High_scaled', 'Pred_Low_scaled'
    ])
    pred_df.sort_values('Date', inplace=True)
    
    # Inverse transform the 4 columns
    scaled_array = pred_df[['Pred_Price_scaled','Pred_Open_scaled','Pred_High_scaled','Pred_Low_scaled']].values
    unscaled_array = inverse_transform_4cols(scaled_array)
    pred_df['Pred_Price_unscaled'] = unscaled_array[:, 0]
    pred_df['Pred_Open_unscaled']  = unscaled_array[:, 1]
    pred_df['Pred_High_unscaled']  = unscaled_array[:, 2]
    pred_df['Pred_Low_unscaled']   = unscaled_array[:, 3]

    return pred_df


# ------------------------------------------------------------------------
# 10. EVALUATE AND PLOT
# ------------------------------------------------------------------------
def evaluate_and_plot(model_name, val_pred_df, test_pred_df):
    """
    Given the val and test predictions (both scaled and unscaled),
    merge with the real val/test sets (scaled) to measure MAPE on Price,
    then do plots (full data plot + zoomed).
    """
    # 1. Merge predictions with real data in val & test for comparison
    df_val_eval = pd.merge(
        df_val_scaled[['Date','Price','Open','High','Low']],
        val_pred_df,
        on='Date', how='left'
    )
    df_test_eval = pd.merge(
        df_test_scaled[['Date','Price','Open','High','Low']],
        test_pred_df,
        on='Date', how='left'
    )

    # Drop rows where we have no predictions
    df_val_eval.dropna(subset=['Pred_Price_scaled'], inplace=True)
    df_test_eval.dropna(subset=['Pred_Price_scaled'], inplace=True)

    # 2. Extract the real scaled Price and predicted scaled Price
    y_val_true_scaled = df_val_eval['Price'].values
    y_val_pred_scaled = df_val_eval['Pred_Price_scaled'].values

    y_test_true_scaled = df_test_eval['Price'].values
    y_test_pred_scaled = df_test_eval['Pred_Price_scaled'].values

    # 3. Inverse transform them to get unscaled Price
    y_val_true_unscaled = extract_price_col( inverse_transform_4cols(df_val_eval[['Price','Open','High','Low']].values) )
    y_val_pred_unscaled = df_val_eval['Pred_Price_unscaled'].values

    y_test_true_unscaled = extract_price_col( inverse_transform_4cols(df_test_eval[['Price','Open','High','Low']].values) )
    y_test_pred_unscaled = df_test_eval['Pred_Price_unscaled'].values

    # 4. Compute MAPE on Price
    val_mape  = mean_absolute_percentage_error(y_val_true_unscaled,  y_val_pred_unscaled)
    test_mape = mean_absolute_percentage_error(y_test_true_unscaled, y_test_pred_unscaled)
    
    val_acc  = 100 - val_mape * 100
    test_acc = 100 - test_mape * 100

    print(f"\n[{model_name}] Validation MAPE on Price: {val_mape * 100:.2f}%, Accuracy: {val_acc:.2f}%")
    print(f"[{model_name}] Test MAPE on Price: {test_mape * 100:.2f}%, Accuracy: {test_acc:.2f}%")

    # 5. Create a combined DataFrame for plotting over entire date range
    #    We'll re-use the walk-forward approach results (val + test),
    #    then compare with the real entire df for Price.
    #    Since we updated the val/test sets in separate calls, we combine them.
    comb_pred = pd.concat([val_pred_df, test_pred_df], axis=0).drop_duplicates(subset=['Date'])
    comb_pred.sort_values('Date', inplace=True)

    # Real entire Price
    real_price_scaled = df_scaled[['Date','Price']].copy()
    real_price_scaled['Price_unscaled'] = extract_price_col(
        inverse_transform_4cols(
            df_scaled[['Price','Open','High','Low']].values
        )
    )
    
    # Merge real with predictions
    merged = pd.merge(real_price_scaled, comb_pred, on='Date', how='left')
    # For days outside val/test, Pred_Price_unscaled is NaN
    # We'll keep them just to show the full timeline.

    # 6. Full Plot
    plt.figure(figsize=(14, 7))
    plt.plot(
        merged['Date'],
        merged['Price_unscaled'],
        color='black',
        label='Actual Price'
    )
    # Overplot predicted
    plt.plot(
        merged['Date'],
        merged['Pred_Price_unscaled'],
        color='red',
        linestyle='-',
        label=f'{model_name} Prediction'
    )

    # Highlight val/test
    plt.axvspan(df_val['Date'].min(), df_val['Date'].max(), color='green', alpha=0.1, label='Val Period')
    plt.axvspan(df_test['Date'].min(), df_test['Date'].max(), color='yellow', alpha=0.1, label='Test Period')

    plt.title(f"[{model_name}] Full Data: Walk-Forward Predictions vs. Actual")
    plt.xlabel("Date")
    plt.ylabel("Gold Price (unscaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 7. Zoomed-in plot for validation period
    zoom_val = merged[(merged['Date'] >= df_val['Date'].min()) & (merged['Date'] <= df_val['Date'].max())]

    plt.figure(figsize=(12, 5))
    plt.plot(zoom_val['Date'], zoom_val['Price_unscaled'], color='black', label='Actual Price')
    plt.plot(zoom_val['Date'], zoom_val['Pred_Price_unscaled'], color='red', linestyle='--', label='Predicted')
    plt.title(f"[{model_name}] Zoomed: Validation Period")
    plt.xlabel("Date")
    plt.ylabel("Price (unscaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 8. Zoomed-in plot for test period
    zoom_test = merged[merged['Date'] >= df_test['Date'].min()]

    plt.figure(figsize=(12, 5))
    plt.plot(zoom_test['Date'], zoom_test['Price_unscaled'], color='black', label='Actual Price')
    plt.plot(zoom_test['Date'], zoom_test['Pred_Price_unscaled'], color='red', linestyle='--', label='Predicted')
    plt.title(f"[{model_name}] Zoomed: Test Period")
    plt.xlabel("Date")
    plt.ylabel("Price (unscaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Return final MAPE values in case we want to store them
    return val_mape, test_mape


# ------------------------------------------------------------------------
# 11. MAIN SCRIPT: TRAIN MULTIPLE MODELS, DO WALK-FORWARD (VAL+TEST), PLOT
# ------------------------------------------------------------------------
if __name__ == "__main__":

    model_types = ['LSTM', 'GRU', 'RNN', 'CNN', 'EnhancedLSTM']
    results_dict = {}

    for mt in model_types:
        print(f"\nTraining model: {mt}")
        model = train_model(model_type=mt)

        # 1) Walk-forward on validation range
        #    Start at the earliest validation date, end at the latest validation date
        val_pred_df = walk_forward_predict(
            model, 
            df_all_scaled=df_scaled,
            start_date=df_val['Date'].min(), 
            end_date=df_val['Date'].max(),
            window=window_width
        )

        # 2) Walk-forward on test range
        test_pred_df = walk_forward_predict(
            model, 
            df_all_scaled=df_scaled,
            start_date=df_test['Date'].min(),
            end_date=df_test['Date'].max(),
            window=window_width
        )

        # 3) Evaluate and plot
        val_mape, test_mape = evaluate_and_plot(mt, val_pred_df, test_pred_df)

        results_dict[mt] = {
            'val_mape': val_mape,
            'test_mape': test_mape
        }

    # Print summary
    print("\nSummary of MAPE on Price:")
    for mt in model_types:
        print(f"{mt}: "
              f"Val MAPE={results_dict[mt]['val_mape']*100:.2f}%, "
              f"Test MAPE={results_dict[mt]['test_mape']*100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F  # for activation functions
from torch.utils.data import DataLoader, TensorDataset
import os
import random
import warnings
import matplotlib.dates as mdates

# -------------------------------
# 1. Set random seeds for reproducibility
# -------------------------------
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(0)

# -------------------------------
# 2. Load and Prepare Data
# -------------------------------
df = pd.read_csv('Gold Futures Historical Data_Original.csv')
df.drop(['Vol.', 'Change %'], axis=1, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)

# Convert numeric columns (remove commas and cast to float)
num_cols = df.columns.drop(['Date'])
df[num_cols] = df[num_cols].replace({',': ''}, regex=True)
df[num_cols] = df[num_cols].astype('float64')

assert df.duplicated().sum() == 0, "Duplicate rows found!"
assert df.isnull().sum().sum() == 0, "Missing values found!"

# We use four features in this example.
features = ['Price', 'Open', 'High', 'Low']
data = df[features].values

# -------------------------------
# 3. Scaling (Avoiding Data Leakage)
# -------------------------------
# Fit the scaler only on training data.
train_mask_for_scaler = df['Date'] < pd.to_datetime('2018-10-24')
data_train = df.loc[train_mask_for_scaler, features].values

scaler = MinMaxScaler()
scaler.fit(data_train)
data_scaled = scaler.transform(data)

# -------------------------------
# 4. Create Sliding Windows (Sequence Preparation)
# -------------------------------
window_width = 30  # length of the input sequence

X, y, dates = [], [], []
for i in range(window_width, len(data_scaled)):
    X.append(data_scaled[i - window_width:i])
    y.append(data_scaled[i, 0])  # target: 'Price' (first column)
    dates.append(df['Date'].iloc[i])
X = np.array(X)
y = np.array(y)
dates = np.array(dates)

assert len(dates) == len(y), "Mismatch between dates and target lengths"

# -------------------------------
# 5. Split Data by Date into Training, Validation and Testing Sets
# -------------------------------
train_mask = dates < pd.to_datetime('2023-10-24')
val_mask = (dates >= pd.to_datetime('2023-10-24')) & (dates < pd.to_datetime('2024-11-28'))
test_mask = dates >= pd.to_datetime('2024-11-28')

X_train, y_train = X[train_mask], y[train_mask]
X_val, y_val = X[val_mask], y[val_mask]
X_test, y_test = X[test_mask], y[test_mask]

train_dates = dates[train_mask]
val_dates = dates[val_mask]
test_dates = dates[test_mask]

# -------------------------------
# 6. Convert Training and Validation Sets to PyTorch Tensors
# -------------------------------
X_train_tensor = torch.from_numpy(X_train).float()
y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1)
X_val_tensor = torch.from_numpy(X_val).float()
y_val_tensor = torch.from_numpy(y_val).float().unsqueeze(1)

# Create TensorDatasets and DataLoaders (for training and validation only)
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

batch_size = 32  # adjust as needed
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)  # no shuffling for time series
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# -------------------------------
# 7. Device Configuration
# -------------------------------
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# -------------------------------
# 8. Model Definitions
# -------------------------------
class BaseModel(nn.Module):
    def __init__(self, model_type='LSTM'):
        super(BaseModel, self).__init__()
        self.model_type = model_type
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                dropout=0.1
            )
            self.fc = nn.Linear(128, 1)
        elif model_type == 'GRU':
            self.rnn = nn.GRU(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                dropout=0.1
            )
            self.fc = nn.Linear(128, 1)
        elif model_type == 'RNN':
            self.rnn = nn.RNN(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                nonlinearity='relu',
                dropout=0.1
            )
            self.fc = nn.Linear(128, 1)
        elif model_type == 'CNN':
            self.conv1 = nn.Conv1d(in_channels=len(features), out_channels=64, kernel_size=3)
            self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3)
            # Two Conv1d layers reduce the sequence length by 2*(kernel_size-1)
            conv_output_length = window_width - 2 * (3 - 1)
            self.fc = nn.Linear(128 * conv_output_length, 1)
        elif model_type == 'EnhancedLSTM':
            self.rnn = nn.LSTM(
                input_size=len(features),
                hidden_size=128,
                num_layers=3,
                batch_first=True,
                dropout=0.2
            )
            self.dropout = nn.Dropout(0.2)
            self.bn = nn.BatchNorm1d(128)
            self.fc = nn.Linear(128, 1)
        else:
            raise ValueError("Invalid model_type. Choose from 'LSTM', 'GRU', 'RNN', 'CNN', or 'EnhancedLSTM'.")
    
    def forward(self, x):
        if self.model_type == 'CNN':
            x = x.permute(0, 2, 1)  # reshape to (batch, channels, seq_length)
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(x.size(0), -1)
            out = self.fc(x)
        elif self.model_type == 'EnhancedLSTM':
            out, _ = self.rnn(x)
            out = out[:, -1, :]
            out = self.bn(out)
            out = self.dropout(out)
            out = self.fc(out)
        else:
            out, _ = self.rnn(x)
            out = out[:, -1, :]
            out = self.fc(out)
        return out

# Early stopping helps us avoid overfitting.
class EarlyStopping:
    def __init__(self, patience=150, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
        
    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

# -------------------------------
# 9. Helper Functions
# -------------------------------
def inverse_transform(y_scaled):
    """
    Inverse-transform the scaled price values back to the original scale.
    y_scaled: array-like, shape (n, ) or (n,1)
    """
    y_scaled = np.array(y_scaled).reshape(-1, 1)
    # Create dummy columns for the remaining features
    zeros = np.zeros((y_scaled.shape[0], len(features) - 1))
    y_combined = np.hstack((y_scaled, zeros))
    y_inv = scaler.inverse_transform(y_combined)
    return y_inv[:, 0]

# Training function (returns the best model and loss histories)
def train_model(model, train_loader, val_loader, best_model_path, num_epochs=800, patience=150):
    model = model.to(device)
    criterion = nn.HuberLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.0005)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=patience)
    early_stopping = EarlyStopping(patience=patience, min_delta=0)
    
    best_val_loss = float('inf')
    train_losses_per_epoch = []
    val_losses_per_epoch = []
    
    for epoch in range(num_epochs):
        model.train()
        train_losses = []
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            output = model(X_batch)
            loss = criterion(output, y_batch)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
        avg_train_loss = np.mean(train_losses)
        train_losses_per_epoch.append(avg_train_loss)
        
        # Validation phase
        model.eval()
        val_losses = []
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                val_output = model(X_batch)
                val_loss = criterion(val_output, y_batch)
                val_losses.append(val_loss.item())
        avg_val_loss = np.mean(val_losses)
        val_losses_per_epoch.append(avg_val_loss)
        
        scheduler.step(avg_val_loss)
        early_stopping(avg_val_loss)
        if early_stopping.early_stop:
            print(f"Early stopping triggered at epoch {epoch+1}")
            break
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), best_model_path)
        
        if (epoch+1) % 100 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")
    
    # Load the best model
    model.load_state_dict(torch.load(best_model_path, map_location=device))
    return model, train_losses_per_epoch, val_losses_per_epoch

# Validation function (non-iterative; uses the actual windows)
def validate_model(model, val_loader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(device)
            y_pred = model(X_batch)
            predictions.append(y_pred.cpu().numpy())
            actuals.append(y_batch.cpu().numpy())
    predictions = np.concatenate(predictions, axis=0)
    actuals = np.concatenate(actuals, axis=0)
    return predictions, actuals

# Iterative testing function (rolling forecast)
def iterative_test(model, initial_window, horizon, num_features):
    """
    Given an initial window (of shape [window_width, num_features]) and a horizon (number of days),
    predict iteratively where each new prediction is appended to update the window.
    Assumption: For the new day, we fill all features with the predicted price.
    """
    model.eval()
    iterative_predictions = []
    current_window = initial_window.copy()  # shape: (window_width, num_features)
    for i in range(horizon):
        # Prepare the input tensor: shape [1, window_width, num_features]
        input_tensor = torch.from_numpy(current_window).float().unsqueeze(0).to(device)
        with torch.no_grad():
            y_pred = model(input_tensor)
        pred_value = y_pred.cpu().item()  # scalar (scaled)
        iterative_predictions.append(pred_value)
        # Create a new row by replicating the predicted price across all features
        new_row = np.full((1, num_features), pred_value)
        # Update the current window: drop the first row, append the new row
        current_window = np.concatenate((current_window[1:], new_row), axis=0)
    return np.array(iterative_predictions)

# -------------------------------
# 10. Training, Validation, and Iterative Testing for Each Model
# -------------------------------
model_names = ['LSTM', 'GRU', 'RNN', 'CNN', 'EnhancedLSTM']
all_results = {}

for model_name in model_names:
    print(f"\nTraining {model_name} model...")
    best_model_path = f'best_model_{model_name}.pt'
    model = BaseModel(model_type=model_name)
    # Train (the function returns the best model and loss histories)
    model, train_losses, val_losses = train_model(model, train_loader, val_loader, best_model_path)
    
    # ---------------------------
    # Validation predictions (using actual past data)
    # ---------------------------
    val_preds_scaled, val_actuals_scaled = validate_model(model, val_loader)
    val_preds_inv = inverse_transform(val_preds_scaled)
    val_actuals_inv = inverse_transform(val_actuals_scaled)
    
    # ---------------------------
    # Iterative Test Predictions
    # For iterative forecasting we use the first test window (from X_test) as the initial input.
    # Then for each future day, the window is updated with the predicted price.
    # ---------------------------
    initial_window = X_test[0]  # shape: (window_width, num_features)
    horizon = len(y_test)      # number of days to forecast in test set
    test_preds_iterative_scaled = iterative_test(model, initial_window, horizon, num_features=len(features))
    test_preds_iterative_inv = inverse_transform(test_preds_iterative_scaled)
    
    # Store results
    all_results[model_name] = {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_dates': val_dates,
        'test_dates': test_dates,
        'val_preds_inv': val_preds_inv,
        'val_actuals_inv': val_actuals_inv,
        'test_preds_iterative_inv': test_preds_iterative_inv,
        'y_test_inv': inverse_transform(y_test)  # actual test prices (scaled back)
    }
    
    # Print performance metrics
    val_mape = mean_absolute_percentage_error(val_actuals_inv, val_preds_inv)
    print(f"[{model_name}] Validation MAPE: {val_mape * 100:.2f}%")
    test_mape = mean_absolute_percentage_error(all_results[model_name]['y_test_inv'], test_preds_iterative_inv)
    print(f"[{model_name}] Iterative Test MAPE: {test_mape * 100:.2f}%")

# -------------------------------
# 11. Plotting the Results
# -------------------------------
# Plot training and validation loss curves for each model
plt.figure(figsize=(14, 7))
for model_name, res in all_results.items():
    plt.plot(res['train_losses'], label=f'{model_name} Train Loss')
    plt.plot(res['val_losses'], label=f'{model_name} Val Loss', linestyle='--')
plt.title("Training and Validation Loss Curves")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot validation predictions vs actual prices
plt.figure(figsize=(14, 7))
plt.plot(val_dates, all_results[model_names[0]]['val_actuals_inv'], color='black', label='Actual Price (Validation)')
for model_name, res in all_results.items():
    plt.plot(res['val_dates'], res['val_preds_inv'], label=f'{model_name} Prediction')
plt.title("Validation Set: Predictions vs Actual Prices")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot iterative test predictions vs actual prices
plt.figure(figsize=(14, 7))
plt.plot(test_dates, all_results[model_names[0]]['y_test_inv'], color='black', label='Actual Price (Test)')
for model_name, res in all_results.items():
    plt.plot(res['test_dates'], res['test_preds_iterative_inv'], label=f'{model_name} Iterative Prediction')
plt.title("Iterative Test Forecasts vs Actual Prices")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F  # for activation functions
from torch.utils.data import DataLoader, TensorDataset
import os
import random
import warnings
import matplotlib.dates as mdates

# =============================================================================
# 1. Set random seeds for reproducibility
# =============================================================================
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(0)

# =============================================================================
# 2. Load and Prepare Data
# =============================================================================
df = pd.read_csv('Gold Futures Historical Data_Original.csv')
df.drop(['Vol.', 'Change %'], axis=1, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)

# Remove commas and convert numeric columns to float
num_cols = df.columns.drop(['Date'])
df[num_cols] = df[num_cols].replace({',': ''}, regex=True)
df[num_cols] = df[num_cols].astype('float64')

assert df.duplicated().sum() == 0, "Duplicate rows found!"
assert df.isnull().sum().sum() == 0, "Missing values found!"

# Use four features: Price, Open, High, Low
features = ['Price', 'Open', 'High', 'Low']
data = df[features].values

# =============================================================================
# 3. Scaling (Avoid Data Leakage)
# =============================================================================
# Fit the scaler only on training data (dates before 2018-10-24)
train_mask_for_scaler = df['Date'] < pd.to_datetime('2023-10-24')
data_train = df.loc[train_mask_for_scaler, features].values

scaler = MinMaxScaler()
scaler.fit(data_train)
data_scaled = scaler.transform(data)

# =============================================================================
# 4. Create Sliding Windows (Sequence Preparation)
# =============================================================================
window_width = 30  # number of past days to use

X, y, dates = [], [], []
# For each day from window_width onward, the input is the previous window_width days
# and the target is the entire 4-feature vector for that day.
for i in range(window_width, len(data_scaled)):
    X.append(data_scaled[i - window_width:i])
    y.append(data_scaled[i, :])  # now predict all four features
    dates.append(df['Date'].iloc[i])
X = np.array(X)  # shape: (samples, window_width, 4)
y = np.array(y)  # shape: (samples, 4)
dates = np.array(dates)

assert len(dates) == len(y), "Mismatch between dates and target lengths"

# =============================================================================
# 5. Split Data by Date into Training, Validation and Testing Sets
# =============================================================================
train_mask  = dates < pd.to_datetime('2023-10-24')
val_mask    = (dates >= pd.to_datetime('2023-10-24')) & (dates < pd.to_datetime('2024-11-28'))
test_mask   = dates >= pd.to_datetime('2024-11-28')

X_train, y_train = X[train_mask], y[train_mask]
X_val, y_val     = X[val_mask], y[val_mask]
X_test, y_test   = X[test_mask], y[test_mask]

train_dates = dates[train_mask]
val_dates   = dates[val_mask]
test_dates  = dates[test_mask]

# =============================================================================
# 6. Convert Training and Validation Sets to PyTorch Tensors
# =============================================================================
X_train_tensor = torch.from_numpy(X_train).float()
y_train_tensor = torch.from_numpy(y_train).float()  # shape: (n, 4)
X_val_tensor   = torch.from_numpy(X_val).float()
y_val_tensor   = torch.from_numpy(y_val).float()

# Create TensorDatasets and DataLoaders (for training and validation only)
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset   = TensorDataset(X_val_tensor, y_val_tensor)

batch_size = 32  # adjust as needed
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)  # no shuffling for time series
val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# =============================================================================
# 7. Device Configuration
# =============================================================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# =============================================================================
# 8. Model Definitions
# =============================================================================
class BaseModel(nn.Module):
    def __init__(self, model_type='LSTM'):
        super(BaseModel, self).__init__()
        self.model_type = model_type
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                dropout=0.1
            )
            self.fc = nn.Linear(128, 4)  # output all 4 features
        elif model_type == 'GRU':
            self.rnn = nn.GRU(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                dropout=0.1
            )
            self.fc = nn.Linear(128, 4)
        elif model_type == 'RNN':
            self.rnn = nn.RNN(
                input_size=len(features),
                hidden_size=128,
                num_layers=2,
                batch_first=True,
                nonlinearity='relu',
                dropout=0.1
            )
            self.fc = nn.Linear(128, 4)
        elif model_type == 'CNN':
            self.conv1 = nn.Conv1d(in_channels=len(features), out_channels=64, kernel_size=3)
            self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3)
            # Two Conv1d layers reduce the sequence length by 2*(kernel_size-1)
            conv_output_length = window_width - 2 * (3 - 1)
            self.fc = nn.Linear(128 * conv_output_length, 4)
        elif model_type == 'EnhancedLSTM':
            self.rnn = nn.LSTM(
                input_size=len(features),
                hidden_size=128,
                num_layers=3,
                batch_first=True,
                dropout=0.2
            )
            self.dropout = nn.Dropout(0.2)
            self.bn = nn.BatchNorm1d(128)
            self.fc = nn.Linear(128, 4)
        else:
            raise ValueError("Invalid model_type. Choose from 'LSTM', 'GRU', 'RNN', 'CNN', or 'EnhancedLSTM'.")
    
    def forward(self, x):
        if self.model_type == 'CNN':
            # For CNN, transpose to (batch, channels, seq_length)
            x = x.permute(0, 2, 1)
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(x.size(0), -1)
            out = self.fc(x)
        elif self.model_type == 'EnhancedLSTM':
            out, _ = self.rnn(x)
            out = out[:, -1, :]  # take last time-step output
            out = self.bn(out)
            out = self.dropout(out)
            out = self.fc(out)
        else:
            out, _ = self.rnn(x)
            out = out[:, -1, :]  # take last time-step output
            out = self.fc(out)
        return out

# Early stopping to avoid overfitting
class EarlyStopping:
    def __init__(self, patience=150, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
        
    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

# =============================================================================
# 9. Helper Functions
# =============================================================================
def inverse_transform(y_scaled):
    """
    Inverse-transform scaled outputs (assumed to be of shape (n,4)) back to original scale.
    """
    y_scaled = np.array(y_scaled)
    if y_scaled.ndim != 2 or y_scaled.shape[1] != len(features):
        raise ValueError("Expected input with shape (n,4)")
    return scaler.inverse_transform(y_scaled)

# Training function (returns the best model and loss histories)
def train_model(model, train_loader, val_loader, best_model_path, num_epochs=500, patience=150):
    model = model.to(device)
    criterion = nn.HuberLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.0005)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=patience)
    early_stopping = EarlyStopping(patience=patience, min_delta=0)
    
    best_val_loss = float('inf')
    train_losses_per_epoch = []
    val_losses_per_epoch = []
    
    for epoch in range(num_epochs):
        model.train()
        train_losses = []
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            output = model(X_batch)
            loss = criterion(output, y_batch)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
        avg_train_loss = np.mean(train_losses)
        train_losses_per_epoch.append(avg_train_loss)
        
        # Validation phase
        model.eval()
        val_losses = []
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                val_output = model(X_batch)
                val_loss = criterion(val_output, y_batch)
                val_losses.append(val_loss.item())
        avg_val_loss = np.mean(val_losses)
        val_losses_per_epoch.append(avg_val_loss)
        
        scheduler.step(avg_val_loss)
        early_stopping(avg_val_loss)
        if early_stopping.early_stop:
            print(f"Early stopping triggered at epoch {epoch+1}")
            break
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), best_model_path)
        
        if (epoch+1) % 100 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")
    
    # Load the best model
    model.load_state_dict(torch.load(best_model_path, map_location=device))
    return model, train_losses_per_epoch, val_losses_per_epoch

# Validation function (non-iterative; uses the actual sliding windows)
def validate_model(model, val_loader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(device)
            y_pred = model(X_batch)
            predictions.append(y_pred.cpu().numpy())
            actuals.append(y_batch.cpu().numpy())
    predictions = np.concatenate(predictions, axis=0)
    actuals = np.concatenate(actuals, axis=0)
    return predictions, actuals

# Iterative testing function (rolling forecast)
def iterative_test(model, initial_window, horizon, num_features):
    """
    Given an initial window (shape: [window_width, num_features]) and a forecast horizon,
    predict iteratively where each new prediction (a vector of 4 values) is appended to update the window.
    """
    model.eval()
    iterative_predictions = []
    current_window = initial_window.copy()  # shape: (window_width, num_features)
    for i in range(horizon):
        input_tensor = torch.from_numpy(current_window).float().unsqueeze(0).to(device)
        with torch.no_grad():
            y_pred = model(input_tensor)  # shape: (1,4)
        pred_vector = y_pred.cpu().numpy()[0]  # shape: (4,)
        iterative_predictions.append(pred_vector)
        # Update the window by dropping the first row and appending the predicted vector
        new_row = pred_vector.reshape(1, num_features)
        current_window = np.concatenate((current_window[1:], new_row), axis=0)
    return np.array(iterative_predictions)  # shape: (horizon, 4)

# =============================================================================
# 10. Training, Validation, and Iterative Testing for Each Model
# =============================================================================
model_names = ['LSTM', 'GRU', 'RNN', 'CNN']
all_results = {}

for model_name in model_names:
    print(f"\nTraining {model_name} model...")
    best_model_path = f'best_model_{model_name}_o3.pt'
    model = BaseModel(model_type=model_name)
    
    # Train the model
    model, train_losses, val_losses = train_model(model, train_loader, val_loader, best_model_path)
    
    # ---------------------------
    # Validation predictions (using actual past windows)
    # ---------------------------
    val_preds_scaled, val_actuals_scaled = validate_model(model, val_loader)
    val_preds_inv   = inverse_transform(val_preds_scaled)    # shape: (n,4)
    val_actuals_inv = inverse_transform(val_actuals_scaled)  # shape: (n,4)
    
    # ---------------------------
    # Iterative Test Predictions (rolling forecast)
    # ---------------------------
    # Use the first test window (from X_test) as the initial input.
    # Forecast for the entire test horizon.
    initial_window = X_test[0]  # shape: (window_width, 4)
    horizon = len(y_test)       # number of days in test set
    test_preds_iterative_scaled = iterative_test(model, initial_window, horizon, num_features=len(features))
    test_preds_iterative_inv = inverse_transform(test_preds_iterative_scaled)
    
    # Also inverse-transform the actual test targets for comparison
    y_test_inv = inverse_transform(y_test)
    
    # Store results
    all_results[model_name] = {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_dates': val_dates,
        'test_dates': test_dates,
        'val_preds_inv': val_preds_inv,           # full 4-feature predictions
        'val_actuals_inv': val_actuals_inv,         # full 4-feature actuals
        'test_preds_iterative_inv': test_preds_iterative_inv,  # full 4-feature iterative predictions
        'y_test_inv': y_test_inv                   # actual test targets (full 4 features)
    }
    
    # Print performance metrics using only the Price (first column)
    val_mape = mean_absolute_percentage_error(val_actuals_inv[:, 0], val_preds_inv[:, 0])
    print(f"[{model_name}] Validation MAPE (Price): {val_mape * 100:.2f}%")
    test_mape = mean_absolute_percentage_error(y_test_inv[:, 0], test_preds_iterative_inv[:, 0])
    print(f"[{model_name}] Iterative Test MAPE (Price): {test_mape * 100:.2f}%")

# =============================================================================
# 11. Plotting the Results
# =============================================================================
# Colors for different models
model_colors = {
    'LSTM': 'red',
    'GRU': 'blue',
    'RNN': 'green',
    'CNN': 'orange',
    'EnhancedLSTM': 'purple',
}

# ---------------------------
# Plot training and validation loss curves for each model
# ---------------------------
plt.figure(figsize=(14, 7))
for model_name, res in all_results.items():
    plt.plot(res['train_losses'], label=f'{model_name} Train Loss')
    plt.plot(res['val_losses'], label=f'{model_name} Val Loss', linestyle='--')
plt.title("Training and Validation Loss Curves")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ---------------------------
# Validation Set: Full Predictions vs. Actual Prices (Price only)
# ---------------------------
plt.figure(figsize=(14, 7))
plt.plot(val_dates, all_results[model_names[0]]['val_actuals_inv'][:, 0], color='black', label='Actual Price (Validation)')
for model_name, res in all_results.items():
    plt.plot(res['val_dates'], res['val_preds_inv'][:, 0], label=f'{model_name} Prediction')
plt.title("Validation Set: Predictions vs Actual Gold Price")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Zoomed-in view for validation period
plt.figure(figsize=(14, 7))
zoom_val_idx = slice(0, min(100, len(val_dates)))  # adjust the slice for zooming as needed
plt.plot(val_dates[zoom_val_idx], all_results[model_names[0]]['val_actuals_inv'][zoom_val_idx, 0], color='black', label='Actual Price (Validation)')
for model_name, res in all_results.items():
    plt.plot(res['val_dates'][zoom_val_idx], res['val_preds_inv'][zoom_val_idx, 0], label=f'{model_name} Prediction')
plt.title("Zoomed-In Validation Set: Gold Price Predictions")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ---------------------------
# Test Set: Iterative Predictions vs. Actual Prices (Price only)
# ---------------------------
plt.figure(figsize=(14, 7))
plt.plot(test_dates, all_results[model_names[0]]['y_test_inv'][:, 0], color='black', label='Actual Price (Test)')
for model_name, res in all_results.items():
    plt.plot(res['test_dates'], res['test_preds_iterative_inv'][:, 0], label=f'{model_name} Iterative Prediction')
plt.title("Test Set: Iterative Forecasts vs Actual Gold Price")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Zoomed-in view for test period
plt.figure(figsize=(14, 7))
zoom_test_idx = slice(0, min(100, len(test_dates)))  # adjust as needed for zoom
plt.plot(test_dates[zoom_test_idx], all_results[model_names[0]]['y_test_inv'][zoom_test_idx, 0], color='black', label='Actual Price (Test)')
for model_name, res in all_results.items():
    plt.plot(res['test_dates'][zoom_test_idx], res['test_preds_iterative_inv'][zoom_test_idx, 0], label=f'{model_name} Iterative Prediction')
plt.title("Zoomed-In Test Set: Iterative Gold Price Forecasts")
plt.xlabel("Date")
plt.ylabel("Gold Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import os
import random
import warnings

# ------------------------------------------------------------------------
# 1. SET RANDOM SEEDS
# ------------------------------------------------------------------------
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(0)

# ------------------------------------------------------------------------
# 2. LOAD AND PREPARE DATA
# ------------------------------------------------------------------------
df = pd.read_csv('Gold Futures Historical Data_Original.csv')
df.drop(['Vol.', 'Change %'], axis=1, inplace=True)

df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)

num_cols = df.columns.drop('Date')
df[num_cols] = df[num_cols].replace({',': ''}, regex=True).astype('float64')

# Basic checks
assert df.duplicated().sum() == 0, "Duplicate rows found!"
assert df.isnull().sum().sum() == 0, "Missing values found!"

features = ['Price', 'Open', 'High', 'Low']
train_cutoff = pd.to_datetime('2024-10-24')
val_cutoff   = pd.to_datetime('2024-12-15')

df_train = df[df['Date'] < train_cutoff]
df_val   = df[(df['Date'] >= train_cutoff) & (df['Date'] < val_cutoff)]
df_test  = df[df['Date'] >= val_cutoff]

# ------------------------------------------------------------------------
# 3. SCALING ON TRAIN ONLY
# ------------------------------------------------------------------------
scaler = MinMaxScaler()
scaler.fit(df_train[features])

df_train_scaled = df_train.copy()
df_val_scaled   = df_val.copy()
df_test_scaled  = df_test.copy()

df_train_scaled[features] = scaler.transform(df_train[features])
df_val_scaled[features]   = scaler.transform(df_val[features])
df_test_scaled[features]  = scaler.transform(df_test[features])

df_scaled = pd.concat([df_train_scaled, df_val_scaled, df_test_scaled], ignore_index=True)

# ------------------------------------------------------------------------
# 4. BUILD (X, y) FOR TRAINING (and VAL) — Single-step, Multi-output
# ------------------------------------------------------------------------
window_width = 30

def make_sequences(df_subset, window=30):
    arr  = df_subset[features].values  # shape [N, 4]
    dts  = df_subset['Date'].values    # shape [N,]
    
    X_list, y_list, date_list = [], [], []
    for i in range(window, len(arr)):
        X_window = arr[i - window : i]   # shape [30,4]
        y_target = arr[i]               # shape [4, ] (Price,Open,High,Low)
        X_list.append(X_window)
        y_list.append(y_target)
        date_list.append(dts[i])
    return np.array(X_list), np.array(y_list), np.array(date_list)

X_train_all, y_train_all, train_dates_all = make_sequences(df_train_scaled, window_width)
X_val_all,   y_val_all,   val_dates_all   = make_sequences(df_val_scaled,   window_width)

X_train_tensor = torch.from_numpy(X_train_all).float()  # [N_train,30,4]
y_train_tensor = torch.from_numpy(y_train_all).float()  # [N_train,4]

X_val_tensor   = torch.from_numpy(X_val_all).float()    # [N_val,30,4]
y_val_tensor   = torch.from_numpy(y_val_all).float()    # [N_val,4]

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset   = TensorDataset(X_val_tensor,   y_val_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False)
val_loader   = DataLoader(val_dataset,   batch_size=32, shuffle=False)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ------------------------------------------------------------------------
# 5. MODEL DEFINITIONS (Multi-output)
# ------------------------------------------------------------------------
class BaseModel(nn.Module):
    def __init__(self, model_type='LSTM'):
        super().__init__()
        self.model_type = model_type
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=2,
                               batch_first=True, dropout=0.1)
            self.fc = nn.Linear(128, 4)
        elif model_type == 'GRU':
            self.rnn = nn.GRU(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, dropout=0.1)
            self.fc = nn.Linear(128, 4)
        elif model_type == 'RNN':
            self.rnn = nn.RNN(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, nonlinearity='relu', dropout=0.1)
            self.fc = nn.Linear(128, 4)
        elif model_type == 'CNN':
            self.conv1 = nn.Conv1d(in_channels=4, out_channels=64, kernel_size=3)
            self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3)
            # Each conv reduces sequence length by (kernel_size-1)
            conv_out_len = window_width - 2*(3 - 1)
            self.fc = nn.Linear(128 * conv_out_len, 4)
        elif model_type == 'EnhancedLSTM':
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=3,
                               batch_first=True, dropout=0.2)
            self.bn = nn.BatchNorm1d(128)
            self.dropout = nn.Dropout(0.2)
            self.fc = nn.Linear(128, 4)
        else:
            raise ValueError("Invalid model_type")
    
    def forward(self, x):
        if self.model_type == 'CNN':
            x = x.permute(0, 2, 1)  # (batch, 4, seq_len)
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(x.size(0), -1)
            out = self.fc(x)
        elif self.model_type == 'EnhancedLSTM':
            out, _ = self.rnn(x)       # [batch,seq,hidden]
            out = out[:, -1, :]       # final step
            out = self.bn(out)
            out = self.dropout(out)
            out = self.fc(out)        # [batch,4]
        else:
            out, _ = self.rnn(x)
            out = out[:, -1, :]       # [batch, hidden]
            out = self.fc(out)        # [batch, 4]
        return out


class EarlyStopping:
    def __init__(self, patience=150, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
    
    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

# ------------------------------------------------------------------------
# 6. INVERSE TRANSFORMS
#    We are predicting 4 columns. We'll invert them all at once.
# ------------------------------------------------------------------------
def inverse_transform_4cols(y_4):
    """
    y_4: (N,4) array of scaled [Price,Open,High,Low].
    Returns unscaled shape [N,4].
    """
    return scaler.inverse_transform(y_4)

def extract_price_col(y_4):
    """
    From array shape [N,4], return the Price column (index=0).
    """
    return y_4[:, 0]

# ------------------------------------------------------------------------
# 7. TRAIN MODEL
# ------------------------------------------------------------------------
def train_model(model_type='LSTM'):
    model = BaseModel(model_type).to(device)
    criterion = nn.MSELoss()  # or nn.HuberLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.0005)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=150)
    early_stopping = EarlyStopping(patience=150, min_delta=0)

    num_epochs = 500
    best_val_loss = float('inf')
    best_path = f'best_model_{model_type}.pt'

    if os.path.exists(best_path):
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=FutureWarning)
            model.load_state_dict(torch.load(best_path, map_location=device))
        print(f"Loaded existing model for {model_type} from {best_path}")

    for epoch in range(num_epochs):
        # TRAIN
        model.train()
        train_losses = []
        for Xb, yb in train_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out = model(Xb)  # [batch,4]
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
        avg_train_loss = np.mean(train_losses)

        # VALIDATION
        model.eval()
        val_losses = []
        with torch.no_grad():
            for Xb, yb in val_loader:
                Xb, yb = Xb.to(device), yb.to(device)
                val_out = model(Xb)
                val_loss = criterion(val_out, yb)
                val_losses.append(val_loss.item())
        avg_val_loss = np.mean(val_losses)

        scheduler.step(avg_val_loss)
        early_stopping(avg_val_loss)
        
        # Save best
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), best_path)

        if early_stopping.early_stop:
            print(f"[{model_type}] Early stopping at epoch {epoch+1}")
            break

        if (epoch + 1) % 100 == 0:
            print(f"[{model_type}] Epoch {epoch+1}/{num_epochs}, "
                  f"Train Loss={avg_train_loss:.6f}, Val Loss={avg_val_loss:.6f}")

    # Load best model
    model.load_state_dict(torch.load(best_path, map_location=device))
    return model

# ------------------------------------------------------------------------
# 8. SINGLE WALK-FORWARD FROM VALIDATION START -> END OF TEST
#    Overwrite predictions in df_work for any day >= val_cutoff
# ------------------------------------------------------------------------
def walk_forward_predict_val_test(model, df_all_scaled, val_start, test_end, window=30):
    """
    model: trained, frozen multi-output model
    df_all_scaled: entire scaled DataFrame
    val_start: earliest validation date
    test_end: last date of the test period

    We do a single pass from val_start to test_end
    Overwriting Price/Open/High/Low for every day in [val_start, test_end].
    
    Returns a DataFrame with columns:
       Date,
       Pred_Price_scaled, Pred_Open_scaled, Pred_High_scaled, Pred_Low_scaled,
       Pred_Price_unscaled, ...
    """
    df_work = df_all_scaled.copy()
    df_work.sort_values('Date', inplace=True)

    # We'll gather predictions in a list
    preds = []

    # All dates from val_start to test_end
    all_dates = df_work[(df_work['Date'] >= val_start) & (df_work['Date'] <= test_end)]['Date'].unique()
    all_dates = np.sort(all_dates)

    model.eval()
    with torch.no_grad():
        for d in all_dates:
            idx_current = df_work.index[df_work['Date'] == d]
            if len(idx_current) == 0:
                continue
            idx_current = idx_current[0]
            start_idx = idx_current - window
            if start_idx < 0:
                # Not enough data for a 30-day window
                continue
            
            # Build the input window from df_work, which may already have predicted values for prior val/test days
            window_data = df_work.iloc[start_idx:idx_current][features].values  # shape [30,4]
            X_input = torch.from_numpy(window_data).float().unsqueeze(0).to(device)  # [1,30,4]

            pred_4_scaled = model(X_input).cpu().numpy()[0]  # shape [4]
            # Overwrite the scaled features for date d
            df_work.loc[idx_current, features] = pred_4_scaled

            preds.append((d, *pred_4_scaled))

    pred_df = pd.DataFrame(preds, columns=[
        'Date',
        'Pred_Price_scaled','Pred_Open_scaled','Pred_High_scaled','Pred_Low_scaled'
    ])
    pred_df.sort_values('Date', inplace=True)

    # Inverse transform
    scaled_arr = pred_df[['Pred_Price_scaled','Pred_Open_scaled','Pred_High_scaled','Pred_Low_scaled']].values
    unscaled_arr = inverse_transform_4cols(scaled_arr)
    pred_df['Pred_Price_unscaled'] = unscaled_arr[:, 0]
    pred_df['Pred_Open_unscaled']  = unscaled_arr[:, 1]
    pred_df['Pred_High_unscaled']  = unscaled_arr[:, 2]
    pred_df['Pred_Low_unscaled']   = unscaled_arr[:, 3]

    return pred_df

# ------------------------------------------------------------------------
# 9. EVALUATION ON VAL AND TEST SLICES (MAPE on Price)
# ------------------------------------------------------------------------
def evaluate_predictions(pred_df):
    """
    pred_df: DataFrame with columns:
       Date, Pred_*_scaled, Pred_*_unscaled, etc.
    We'll merge with df_scaled to get the real scaled data,
    then slice for val, test, compute MAPE on Price.
    Returns dict with stats & also a merged DataFrame for plotting.
    """
    # Merge with real scaled data to find ground-truth
    # df_scaled has columns Date + [Price,Open,High,Low]
    merged = pd.merge(
        df_scaled[['Date','Price','Open','High','Low']], 
        pred_df, 
        on='Date', how='left'
    )
    merged.sort_values('Date', inplace=True)

    # Inverse transform real scaled [Price,Open,High,Low]
    # so we can get actual unscaled Price
    real_4_scaled = merged[['Price','Open','High','Low']].values
    real_4_unscaled = inverse_transform_4cols(real_4_scaled)
    merged['Real_Price_unscaled'] = real_4_unscaled[:,0]
    merged['Real_Open_unscaled']  = real_4_unscaled[:,1]
    merged['Real_High_unscaled']  = real_4_unscaled[:,2]
    merged['Real_Low_unscaled']   = real_4_unscaled[:,3]

    # We'll define date-based slices for val & test
    # validation: [train_cutoff, val_cutoff) -> We want [val_cutoff_start, test_cutoff)
    val_mask  = (merged['Date'] >= train_cutoff) & (merged['Date'] <  val_cutoff)
    test_mask = (merged['Date'] >= val_cutoff)

    # We only compute metrics where we actually have predictions
    has_pred_mask = ~merged['Pred_Price_scaled'].isna()

    # Validation slice
    val_slice  = merged[val_mask & has_pred_mask]
    # Test slice
    test_slice = merged[test_mask & has_pred_mask]

    # MAPE on Price
    def compute_mape(slice_df):
        y_true = slice_df['Real_Price_unscaled'].values
        y_pred = slice_df['Pred_Price_unscaled'].values
        if len(y_true) == 0:
            return None
        return mean_absolute_percentage_error(y_true, y_pred)

    val_mape  = compute_mape(val_slice)
    test_mape = compute_mape(test_slice)

    return {
        'val_mape': val_mape,
        'test_mape': test_mape,
        'merged': merged  # for plotting
    }

# ------------------------------------------------------------------------
# 10. COMBINED PLOT FOR ALL MODELS
# ------------------------------------------------------------------------
def plot_all_models(all_results):
    """
    all_results: dict of {model_name: result_dict}, where result_dict['merged'] is
                 the DataFrame with columns Real_Price_unscaled, Pred_Price_unscaled, etc.
    We'll produce:
      - One big plot from the earliest date in merged to the end,
        overplot each model's predicted Price
      - We'll highlight val/test
      - We'll do separate zoom-in for val & test with all model lines
    """
    # 1. Build a single "real" timeline from df_scaled
    df_real = df_scaled[['Date','Price','Open','High','Low']].copy()
    df_real.sort_values('Date', inplace=True)
    real_unscaled = inverse_transform_4cols(df_real[features].values)
    df_real['Real_Price_unscaled'] = real_unscaled[:,0]

    # We'll store each model's predictions in a column for the full timeline
    # We'll do an outer merge on Date so we can align them
    # Start from df_real as a base
    for model_name, info in all_results.items():
        merged = info['merged']
        # We'll rename Pred_Price_unscaled -> e.g. Pred_Price_unscaled_{model_name}
        colname = f'Pred_Price_unscaled_{model_name}'
        # We'll do a left merge on df_real
        tmp = merged[['Date','Pred_Price_unscaled']].rename(columns={'Pred_Price_unscaled': colname})
        df_real = pd.merge(df_real, tmp, on='Date', how='left')

    # 2. Full Plot
    plt.figure(figsize=(14, 7))
    plt.plot(df_real['Date'], df_real['Real_Price_unscaled'], color='black', label='Actual Price')

    colors = {
        'LSTM': 'red',
        'GRU': 'blue',
        'RNN': 'green',
        'CNN': 'orange',
    }
    
    for model_name in all_results.keys():
        colname = f'Pred_Price_unscaled_{model_name}'
        plt.plot(
            df_real['Date'], df_real[colname],
            color=colors.get(model_name, 'gray'),
            linestyle='-',
            label=f'{model_name} Prediction'
        )

    # Highlight val/test
    plt.axvspan(train_cutoff, val_cutoff, color='green', alpha=0.1, label='Validation Period')
    plt.axvspan(val_cutoff, df_real['Date'].max(), color='yellow', alpha=0.1, label='Test Period')

    plt.title("All Models: Full Timeline Predictions vs. Actual")
    plt.xlabel("Date")
    plt.ylabel("Gold Price (unscaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 3. Zoomed-in: Validation period
    df_val_zoom = df_real[(df_real['Date'] >= train_cutoff) & (df_real['Date'] < val_cutoff)]

    plt.figure(figsize=(12, 5))
    plt.plot(df_val_zoom['Date'], df_val_zoom['Real_Price_unscaled'], color='black', label='Actual Price')
    for model_name in all_results.keys():
        colname = f'Pred_Price_unscaled_{model_name}'
        plt.plot(
            df_val_zoom['Date'], df_val_zoom[colname],
            color=colors.get(model_name, 'gray'),
            linestyle='--',
            label=f'{model_name} Prediction'
        )
    plt.title("All Models: Zoomed Validation Period")
    plt.xlabel("Date")
    plt.ylabel("Gold Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 4. Zoomed-in: Test period
    df_test_zoom = df_real[df_real['Date'] >= val_cutoff]

    plt.figure(figsize=(12, 5))
    plt.plot(df_test_zoom['Date'], df_test_zoom['Real_Price_unscaled'], color='black', label='Actual Price')
    for model_name in all_results.keys():
        colname = f'Pred_Price_unscaled_{model_name}'
        plt.plot(
            df_test_zoom['Date'], df_test_zoom[colname],
            color=colors.get(model_name, 'gray'),
            linestyle='--',
            label=f'{model_name} Prediction'
        )
    plt.title("All Models: Zoomed Test Period")
    plt.xlabel("Date")
    plt.ylabel("Gold Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# ------------------------------------------------------------------------
# 11. MAIN: TRAIN MULTIPLE MODELS, SINGLE CONTINUOUS WALK-FORWARD, PLOT
# ------------------------------------------------------------------------
if __name__ == "__main__":
    model_types = ['LSTM', 'GRU', 'RNN', 'CNN']
    all_results = {}

    for mt in model_types:
        print(f"\n--- Training {mt} ---")
        model = train_model(mt)

        # Single walk-forward from earliest validation day (train_cutoff) to the end of test
        # We'll make sure we start walk-forward at the earliest day we consider "blind."
        # That is the day we want to begin overwriting with predictions.
        pred_df = walk_forward_predict_val_test(
            model,
            df_all_scaled=df_scaled,
            val_start=train_cutoff,   # Overwrite from the first val day onward
            test_end=df['Date'].max(),  # up to last date in entire dataset
            window=window_width
        )

        # Evaluate on validation slice [train_cutoff, val_cutoff) and test slice [val_cutoff, end]
        info = evaluate_predictions(pred_df)
        val_mape  = info['val_mape']
        test_mape = info['test_mape']

        if val_mape is None:
            val_mape_pct = "N/A"
        else:
            val_mape_pct = f"{val_mape * 100:.2f}%"

        if test_mape is None:
            test_mape_pct = "N/A"
        else:
            test_mape_pct = f"{test_mape * 100:.2f}%"

        print(f"{mt}: Val MAPE={val_mape_pct}, Test MAPE={test_mape_pct}")

        # Save
        info['pred_df'] = pred_df  # raw predictions table
        all_results[mt] = info

    # Summaries
    print("\nSummary of MAPE on Price:")
    for mt in model_types:
        vmap = all_results[mt]['val_mape']
        tmap = all_results[mt]['test_mape']
        vstr = f"{vmap*100:.2f}%" if vmap is not None else "N/A"
        tstr = f"{tmap*100:.2f}%" if tmap is not None else "N/A"
        print(f"  {mt}:  Val={vstr}, Test={tstr}")

    # Plot all models on one set of charts
    plot_all_models(all_results)
```
