```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
pd.options.mode.chained_assignment = None

from sklearn.preprocessing import MinMaxScaler
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import os
import random
import joblib
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor

# ============================================
# Global seeds & settings
# ============================================
torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(0)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -- Hyperparameters
WINDOW_SIZE = 35
NUM_EPOCHS  = 1000
features = ["Price", "Open", "High", "Low"]

# File paths
csv_file_train    = "Silver Futures Historical Data.csv"
csv_file_complete = "Silver Futures Historical Data_Complete.csv"
best_models_dir   = "."

# ============================================
# 1. Load & Clean CSV (Training)
# ============================================
df = pd.read_csv(csv_file_train)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
for drop_col in ["Vol.", "Change %"]:
    if drop_col in df.columns:
        df.drop(columns=[drop_col], inplace=True, errors="ignore")

for col in features:
    if col not in df.columns:
        df[col] = np.nan
    else:
        df[col] = df[col].astype(str).str.replace(",", "", regex=True)
        df[col] = pd.to_numeric(df[col], errors="coerce")

df.sort_values("Date", inplace=True)
df.reset_index(drop=True, inplace=True)

print("===== TRAINING CSV INFO =====")
print("  Min date:", df["Date"].min())
print("  Max date:", df["Date"].max())
print("Rows:", len(df))

train_cutoff = pd.to_datetime("2024-08-01")
val_cutoff   = pd.to_datetime("2024-12-31")
test_cutoff  = pd.to_datetime("2025-01-01")

df_train = df[df["Date"] < train_cutoff].copy()
df_val   = df[(df["Date"] >= train_cutoff) & (df["Date"] <= val_cutoff)].copy()
df_test  = df[df["Date"] >= test_cutoff].copy()

print(f"TRAIN rows: {len(df_train)}, VAL: {len(df_val)}, TEST: {len(df_test)}")

df_train_nonan = df_train.dropna(subset=features)
if df_train_nonan.empty:
    raise ValueError("No valid numeric feature rows in the training set...")

# ============================================
# 2. Fit MinMaxScaler on TRAIN
# ============================================
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df_train_nonan[features])

df_train_scaled = df_train.copy()
df_val_scaled   = df_val.copy()
df_test_scaled  = df_test.copy()

df_train_scaled.loc[df_train_nonan.index, features] = scaler.transform(df_train_nonan[features])
val_no_nan = df_val_scaled.dropna(subset=features)
if not val_no_nan.empty:
    df_val_scaled.loc[val_no_nan.index, features] = scaler.transform(val_no_nan[features])

train_mins = df_train_nonan[features].min()
df_test_filled = df_test_scaled[features].fillna(train_mins)
df_test_scaled.loc[:, features] = scaler.transform(df_test_filled)

df_scaled = pd.concat([df_train_scaled, df_val_scaled, df_test_scaled], ignore_index=True)
joblib.dump(scaler, "scaler.pkl")

# ============================================
# 3. Sequence creation function
# ============================================
def make_sequences(df_subset, window=35):
    arr = df_subset[features].values
    dts = df_subset["Date"].values
    X_list, y_list, date_list = [], [], []
    for i in range(window, len(arr)):
        X_window = arr[i - window: i]
        y_target = arr[i]
        X_list.append(X_window)
        y_list.append(y_target)
        date_list.append(dts[i])
    return np.array(X_list), np.array(y_list), np.array(date_list)

df_train_nonan = df_train_scaled.dropna(subset=features)
X_train_all, y_train_all, _ = make_sequences(df_train_nonan, WINDOW_SIZE)

# ============================================
# 4. Model Definitions
# ============================================
class BaseModel(nn.Module):
    def __init__(self, model_type="LSTM", window_width=35):
        super().__init__()
        self.model_type = model_type

        if model_type == "CNN":
            # Reduced dropout => simply not used in CNN here
            self.conv1 = nn.Conv1d(in_channels=4, out_channels=64, kernel_size=3)
            self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3)
            with torch.no_grad():
                dummy_input = torch.zeros(1, 4, window_width)
                dummy_output = self.conv2(F.relu(self.conv1(dummy_input)))
                conv_output_size = dummy_output.shape[1] * dummy_output.shape[2]
            self.fc = nn.Linear(conv_output_size, 4)

        elif model_type == "LSTM":
            # Lower dropout
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=2,
                               batch_first=True, dropout=0.05)
            self.fc  = nn.Linear(128, 4)

        elif model_type == "GRU":
            self.rnn = nn.GRU(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, dropout=0.05)
            self.fc  = nn.Linear(128, 4)

        elif model_type == "RNN":
            self.rnn = nn.RNN(input_size=4, hidden_size=128, num_layers=2,
                              batch_first=True, nonlinearity="relu", dropout=0.05)
            self.fc  = nn.Linear(128, 4)

        elif model_type == "EnhancedLSTM":
            # Lower dropout
            self.rnn = nn.LSTM(input_size=4, hidden_size=128, num_layers=3,
                               batch_first=True, dropout=0.05)
            self.bn  = nn.BatchNorm1d(128)
            self.dropout = nn.Dropout(0.05)
            self.fc  = nn.Linear(128, 4)

        elif model_type == "Transformer":
            self.input_linear = nn.Linear(4, 128)
            encoder_layer = nn.TransformerEncoderLayer(d_model=128, nhead=8, dropout=0.05)
            self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=3)
            self.fc = nn.Linear(128, 4)

        elif model_type in ["N-BEATS","N-HITS"]:
            self.input_size = window_width*4
            self.blocks = nn.ModuleList([nn.Sequential(
                nn.Linear(self.input_size, 128),
                nn.ReLU(),
                nn.Linear(128, 128),
                nn.ReLU(),
                nn.Linear(128, 4)
            ) for _ in range(3)])

        else:
            raise ValueError("Invalid model_type")

    def forward(self, x):
        if self.model_type == "CNN":
            x = x.permute(0,2,1)
            x = F.relu(self.conv1(x))
            x = F.relu(self.conv2(x))
            x = x.view(x.size(0), -1)
            return self.fc(x)

        elif self.model_type in ["LSTM","GRU","RNN"]:
            out, _ = self.rnn(x)
            out = out[:, -1, :]
            return self.fc(out)

        elif self.model_type == "EnhancedLSTM":
            out, _ = self.rnn(x)
            out = out[:, -1, :]
            out = self.bn(out)
            out = self.dropout(out)
            return self.fc(out)

        elif self.model_type == "Transformer":
            x = self.input_linear(x)
            x = x.permute(1,0,2)
            x = self.transformer_encoder(x)
            x = x[-1,:,:]
            return self.fc(x)

        elif self.model_type in ["N-BEATS","N-HITS"]:
            x_flat= x.reshape(x.size(0),-1)
            forecast= 0
            for block in self.blocks:
                forecast += block(x_flat)
            return forecast

        else:
            raise ValueError("Invalid model_type")

class EarlyStopping:
    def __init__(self, patience=100, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                print("Early stopping triggered")
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

def inverse_transform_4cols(y_4):
    return scaler.inverse_transform(y_4)

# ============================================
# 5A. Inject random noise into validation data
# ============================================
def add_random_noise(df_in, val_start, val_end, noise_std=0.1):
    """
    In-place add random Gaussian noise to the 'features' columns for rows in [val_start..val_end].
    """
    df_noisy= df_in.copy()
    val_mask= (df_noisy["Date"]>=val_start)&(df_noisy["Date"]<=val_end)
    idxs= df_noisy.index[val_mask]
    if len(idxs)==0:
        return df_noisy
    # Add noise to the relevant features
    for c in features:
        arr_original= df_noisy.loc[idxs, c].values
        noise= np.random.normal(loc=0.0, scale=noise_std, size=len(arr_original))
        df_noisy.loc[idxs, c]= arr_original+ noise
    return df_noisy

# ============================================
# 5B. Validation Loss with noise & L1
# ============================================
def half_blind_validation_loss(model, df_scaled, val_start, val_end, window=35):
    """
    L1Loss + random noise in the validation window => encourages bigger variety.
    """
    criterion = nn.L1Loss()  # <--- alternate loss: MAE
    # copy & inject noise into validation portion
    df_noisy= add_random_noise(df_scaled, val_start, val_end, noise_std=0.1)

    df_noisy.sort_values("Date", inplace=True)
    val_dates= np.sort(df_noisy.loc[(df_noisy["Date"]>=val_start)&(df_noisy["Date"]<=val_end),"Date"].unique())

    model.eval()
    losses=[]
    with torch.no_grad():
        for d in val_dates:
            idx= df_noisy.index[df_noisy["Date"]==d]
            if len(idx)==0:
                continue
            idx= idx[0]
            start_idx= idx- window
            if start_idx<0:
                continue
            window_df= df_noisy.iloc[start_idx:idx][features].values
            if np.isnan(window_df).any():
                continue
            X_in= torch.tensor(window_df,dtype=torch.float).unsqueeze(0).to(device)
            pred_scaled= model(X_in).cpu()[0]
            realvals= df_noisy.loc[idx, features].values.astype(float,copy=False)
            if np.isnan(realvals).any():
                continue
            loss= criterion(pred_scaled, torch.tensor(realvals).float())
            losses.append(loss.item())

    if not losses:
        return 999999.0
    return np.mean(losses)

# ============================================
# 5C. half_blind_validation_preds_df with noise (if you want)
#     or we can keep the old version
# ============================================
def half_blind_validation_preds_df(model, df_scaled, val_start, val_end, window=35):
    """
    For demonstration, we won't add noise here. If you want noise in validation predictions, 
    you can do so similarly. 
    """
    df_work = df_scaled.copy()
    df_work.sort_values("Date", inplace=True)
    val_dates = np.sort(df_work.loc[(df_work["Date"]>=val_start)&(df_work["Date"]<=val_end),"Date"].unique())

    model.eval()
    preds=[]
    with torch.no_grad():
        for d in val_dates:
            idx= df_work.index[df_work["Date"]== d]
            if len(idx)==0:
                continue
            idx= idx[0]
            st= idx- window
            if st<0:
                continue
            arr= df_work.iloc[st:idx][features].values
            if np.isnan(arr).any():
                continue
            X_in= torch.tensor(arr,dtype=torch.float).unsqueeze(0).to(device)
            p_scaled= model(X_in).cpu().numpy()[0]
            preds.append((d,*p_scaled))

    dfp= pd.DataFrame(preds, columns=["Date","Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"])
    dfp.sort_values("Date", inplace=True)
    arr_ = dfp[["Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"]].values
    arr_un= inverse_transform_4cols(arr_)
    dfp["Pred_Price_unscaled"]= arr_un[:,0]
    dfp["Pred_Open_unscaled"]=  arr_un[:,1]
    dfp["Pred_High_unscaled"]= arr_un[:,2]
    dfp["Pred_Low_unscaled"]=  arr_un[:,3]
    return dfp

def iterative_half_blind_validation_preds(model, df_all_scaled, val_start, val_end, window=35):
    df_work = df_all_scaled.copy()
    df_work.sort_values("Date", inplace=True)
    
    # Ensure columns are float
    df_work[features] = df_work[features].apply(pd.to_numeric, errors='coerce')
    df_work[features] = df_work[features].fillna(method='ffill').fillna(method='bfill')
    
    val_idx = df_work.index[df_work["Date"] >= val_start]
    if len(val_idx) == 0:
        return pd.DataFrame()
    first_val_idx = val_idx[0]

    current_window = df_work.iloc[first_val_idx - window:first_val_idx][features].values
    model.eval()
    preds = []
    val_dates = np.sort(df_work[(df_work["Date"] >= val_start)&(df_work["Date"]<= val_end)]["Date"].unique())
    with torch.no_grad():
        for d in val_dates:
            idx = df_work.index[df_work["Date"] == d]
            if len(idx) == 0:
                continue
            idx = idx[0]
            # Convert current_window to a float32 array
            current_window = current_window.astype(np.float32, copy=False)
            X_input = torch.tensor(current_window, dtype=torch.float).unsqueeze(0).to(device)
            pred_scaled = model(X_input).cpu().numpy()[0]
            preds.append((d, *pred_scaled))

            # reveal real row for next iteration
            actual_row = df_work.loc[idx, features].values
            current_window = np.vstack((current_window[1:], actual_row))

    # Convert to DataFrame
    pred_df = pd.DataFrame(preds, columns=["Date","Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"])
    pred_df.sort_values("Date", inplace=True)
    arr_scaled = pred_df[["Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"]].values
    arr_unscaled = inverse_transform_4cols(arr_scaled)
    pred_df["Pred_Price_unscaled"] = arr_unscaled[:,0]
    pred_df["Pred_Open_unscaled"]  = arr_unscaled[:,1]
    pred_df["Pred_High_unscaled"]  = arr_unscaled[:,2]
    pred_df["Pred_Low_unscaled"]   = arr_unscaled[:,3]
    return pred_df


# ============================================
# 6. "Train or Load" with reduced L2 & L1 Loss
# ============================================
class SklearnWrapper:
    def __init__(self, model):
        self.model= model
    def forward(self, x):
        arr= x.cpu().numpy().reshape(1,-1)
        y= self.model.predict(arr)
        return torch.from_numpy(y).float().to(device)
    def eval(self):
        pass
    def __call__(self,x):
        return self.forward(x)

def train_or_load_torch(model_type, 
                        window_size,
                        df_scaled_in,
                        train_cutoff_in,
                        val_cutoff_in,
                        model_save_name="best_model.pt",
                        num_epochs=1000):
    if os.path.exists(model_save_name):
        print(f"[PyTorch] Loading {model_type} from {model_save_name}")
        model= BaseModel(model_type, window_size).to(device)
        model.load_state_dict(torch.load(model_save_name, map_location=device))
        model.eval()
        return model
    else:
        print(f"[PyTorch] No file => Training {model_type} from scratch => {model_save_name}")
        df_work= df_scaled_in.copy()
        df_work.sort_values("Date", inplace=True)

        df_train_local= df_work[df_work["Date"]< train_cutoff_in].copy()
        df_val_local=   df_work[(df_work["Date"]>= train_cutoff_in)&(df_work["Date"]<= val_cutoff_in)].copy()

        df_train_nonan= df_train_local.dropna(subset=features)
        if df_train_nonan.empty:
            raise ValueError("No training data in rolling window")

        X_, y_, _= make_sequences(df_train_nonan, window_size)
        X_t= torch.from_numpy(X_).float()
        y_t= torch.from_numpy(y_).float()
        ds= TensorDataset(X_t,y_t)
        dl= DataLoader(ds, batch_size=len(ds), shuffle=False)

        model= BaseModel(model_type, window_size).to(device)
        # Alternate loss => L1Loss
        criterion= nn.L1Loss()
        # No L2 => weight_decay=0
        optimizer= optim.AdamW(model.parameters(), lr=0.0009, weight_decay=0.0005)

        early_stopper= EarlyStopping(patience=100)
        best_loss= float('inf')

        for epoch in range(num_epochs):
            model.train()
            for Xb, yb in dl:
                Xb, yb= Xb.to(device), yb.to(device)
                optimizer.zero_grad()
                out= model(Xb)
                loss= criterion(out,yb)
                loss.backward()
                optimizer.step()

            # validation
            val_loss= half_blind_validation_loss(
                model,
                df_scaled_in,
                val_start= train_cutoff_in,
                val_end=   val_cutoff_in,
                window= window_size
            )
            if epoch%200==0 and epoch>0:
                print(f"Epoch {epoch}: val_loss={val_loss:.6f}")

            early_stopper(val_loss)
            if val_loss< best_loss:
                best_loss= val_loss
                torch.save(model.state_dict(), model_save_name)
            if early_stopper.early_stop:
                print(f"Early stopping at epoch {epoch}")
                break

        model.load_state_dict(torch.load(model_save_name, map_location=device))
        model.eval()
        return model

def train_or_load_sklearn(model_type,
                          df_scaled_in,
                          train_cutoff_in,
                          val_cutoff_in,
                          model_save_name="best_sklearn.pkl",
                          window_size=35):
    if os.path.exists(model_save_name):
        print(f"[sklearn] Loading {model_type} => {model_save_name}")
        loaded= joblib.load(model_save_name)
        return SklearnWrapper(loaded)
    else:
        print(f"[sklearn] No file => Train {model_type} => {model_save_name}")
        df_work= df_scaled_in.copy()
        df_work.sort_values("Date", inplace=True)

        df_train_local= df_work[df_work["Date"]< train_cutoff_in].copy()
        df_val_local=   df_work[(df_work["Date"]>= train_cutoff_in)&(df_work["Date"]<= val_cutoff_in)].copy()

        df_train_nonan= df_train_local.dropna(subset=features)
        if df_train_nonan.empty:
            raise ValueError("No training data for sklearn in rolling window")

        X_, y_, _= make_sequences(df_train_nonan, window_size)
        Xf= X_.reshape((X_.shape[0], window_size*4))

        if model_type=="SVM":
            base= SVR(kernel='rbf', C=10, epsilon=0.01)
        elif model_type=="GPR":
            base= GaussianProcessRegressor(alpha=1e-2)
        elif model_type=="Boost":
            base= GradientBoostingRegressor(n_estimators=200, max_depth=3)
        else:
            raise ValueError("Unsupported sklearn model")

        mo= MultiOutputRegressor(base)
        mo.fit(Xf,y_)
        joblib.dump(mo, model_save_name)
        return SklearnWrapper(mo)

# ============================================
# 7. Forecast function (full-blind) w/ noise in test
# ============================================
def forecast_n_days_from_date(model, df_actual, start_date, window=35, horizon=30, 
                              device=None, scaler=None, noise_std=0.05):
    """
    Full-blind => feed predicted day into next day. 
    We also add a small random noise to each newly predicted day
    to encourage more varied (less smooth) next step.
    """
    def add_noise_to_pred(pred_unscaled, noise_std):
        # Add small random Gaussian noise to each dimension
        noise= np.random.normal(0.0, noise_std, size=len(pred_unscaled))
        return np.clip(pred_unscaled + noise, a_min=0, a_max=None)  # no negative

    arr_window = get_latest_window(df_actual, start_date, window, scaler)
    if arr_window is None:
        return pd.DataFrame()

    rolling_buffer = np.copy(arr_window)
    forecast_records = []
    current_date = pd.to_datetime(start_date)

    for i in range(horizon):
        X_input = torch.tensor(rolling_buffer, dtype=torch.float).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_scaled = model(X_input).cpu().numpy()[0]

        # Invert scale
        pred_unscaled = scaler.inverse_transform(pred_scaled.reshape(1,-1))[0]
        # Add a bit of random noise to encourage volatility
        pred_noisy = add_noise_to_pred(pred_unscaled, noise_std)

        dayx = current_date + pd.Timedelta(days=i)
        forecast_records.append({
            "ForecastDate": dayx,
            "Pred_Price_unscaled": pred_noisy[0],
            "Pred_Open_unscaled":  pred_noisy[1],
            "Pred_High_unscaled":  pred_noisy[2],
            "Pred_Low_unscaled":   pred_noisy[3],
        })

        # Convert back to scaled and feed forward
        pred_scaled_noisy= scaler.transform(pred_noisy.reshape(1,-1))[0]
        rolling_buffer = np.vstack((rolling_buffer[1:], pred_scaled_noisy.reshape(1,-1)))

    return pd.DataFrame(forecast_records)


def get_latest_window(actual_df, current_date, window=35, scaler=None):
    mask = actual_df["Date"] < current_date
    df_sub = actual_df.loc[mask].copy()
    df_sub.sort_values("Date", inplace=True)
    if len(df_sub) < window:
        return None
    df_window = df_sub.iloc[-window:].copy()
    df_window[features] = df_window[features].fillna(method="ffill").fillna(method="bfill")
    arr = scaler.transform(df_window[features])
    return arr


# ============================================
# 8A. Daily plotting 
# ============================================
def plot_daily_two_charts(day_i, horizon_days, daily_fcst_dict, df_actual):
    df_actual_sorted= df_actual.copy()
    df_actual_sorted.sort_values("Date", inplace=True)

    day_end = day_i + pd.Timedelta(days=horizon_days)
    mask_full= (df_actual_sorted["Date"]<= day_end)
    df_plot_full= df_actual_sorted.loc[mask_full].copy()

    # 1) Joint Diagram
    plt.figure(figsize=(12,5))
    plt.plot(df_plot_full["Date"], df_plot_full["Price"], color="black", label="Actual Price")

    color_map= {
        "LSTM": "red",
        "GRU": "blue",
        "RNN": "green",
        "CNN": "orange",
        "EnhancedLSTM": "magenta",
        "Transformer": "cyan",
        "N-BEATS": "brown",
        "N-HITS": "pink",
        "SVM": "olive",
        "GPR": "teal",
        "Boost": "purple"
    }
    for mname, fcdf in daily_fcst_dict.items():
        subfc= fcdf[fcdf["ForecastDate"]<= day_end].copy()
        if not subfc.empty:
            plt.plot(
                subfc["ForecastDate"],
                subfc["Pred_Price_unscaled"],
                color=color_map.get(mname,"gray"),
                linestyle="--",
                label=f"{mname} Forecast"
            )

    plt.title(f"Day {day_i.date()} - Joint Diagram (up to {day_end.date()})")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 2) Zoom-In
    plt.figure(figsize=(12,5))
    mask_zoom= (df_plot_full["Date"]>= day_i)&(df_plot_full["Date"]<= day_end)
    df_zoom= df_plot_full[mask_zoom].copy()
    if not df_zoom.empty:
        plt.plot(
            df_zoom["Date"],
            df_zoom["Price"],
            color="black",
            label="Actual Price"
        )
    for mname, fcdf in daily_fcst_dict.items():
        subfc= fcdf[(fcdf["ForecastDate"]>= day_i)&(fcdf["ForecastDate"]<= day_end)]
        if not subfc.empty:
            plt.plot(
                subfc["ForecastDate"],
                subfc["Pred_Price_unscaled"],
                color=color_map.get(mname,"gray"),
                linestyle="--",
                label=f"{mname} Forecast"
            )
    plt.title(f"Day {day_i.date()} - Zoom-In Forecast ({day_i.date()} ~ {day_end.date()})")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ============================================
# 8B. Rolling day-by-day
# ============================================
def rolling_train_validate_predict_moe(
    df_full,
    scaler_obj,
    model_types,
    start_train_cutoff = pd.to_datetime("2024-08-01"),
    start_val_cutoff   = pd.to_datetime("2024-12-31"),
    start_pred         = pd.to_datetime("2025-01-02"),
    end_pred           = pd.to_datetime("2025-01-31"),
    horizon_days=30,
    do_daily_plots=True
):
    df_sorted= df_full.copy()
    df_sorted.sort_values("Date", inplace=True)

    df_scaled_all= df_sorted.copy()
    df_scaled_all[features]= df_scaled_all[features].fillna(method="ffill").fillna(method="bfill")
    df_scaled_all[features]= scaler_obj.transform(df_scaled_all[features])

    all_records=[]
    date_range= pd.date_range(start_pred, end_pred, freq="D")

    for i, day_i in enumerate(date_range):
        train_cutoff_i= start_train_cutoff + pd.Timedelta(days=i)
        val_cutoff_i=   start_val_cutoff   + pd.Timedelta(days=i)

        print(f"\n=== Rolling Day {day_i.date()} ===")
        print(f"  Train cutoff={train_cutoff_i.date()}, Val cutoff={val_cutoff_i.date()}")
        print(f"  Horizon={horizon_days} from {day_i.date()} (FULL-BLIND w/ noise)")

        # 1) Train or Load
        model_dict_i= {}
        for mt in model_types:
            out_filename= f"best_{mt}_{day_i.strftime('%Y-%m-%d')}_Mixture_of_Experts_V1"
            if mt in ["SVM","GPR","Boost"]:
                out_filename+= ".pkl"
                m= train_or_load_sklearn(
                    mt, df_scaled_all, train_cutoff_i, val_cutoff_i, out_filename, window_size=WINDOW_SIZE
                )
            else:
                out_filename+= ".pt"
                m= train_or_load_torch(
                    mt, WINDOW_SIZE, df_scaled_all, train_cutoff_i, val_cutoff_i, out_filename, num_epochs=NUM_EPOCHS
                )
            model_dict_i[mt]= m

        # 2) Forecast 30 days
        daily_fcst_dict= {}
        for mt,mobj in model_dict_i.items():
            df_fc= forecast_n_days_from_date(
                mobj,
                df_sorted,
                start_date= day_i,
                window= WINDOW_SIZE,
                horizon= horizon_days,
                device=device,
                scaler=scaler_obj,
                noise_std=0.005  # more/less noise
            )
            if not df_fc.empty:
                df_fc["BaseDate"]= day_i
                df_fc["Model"]= mt
                daily_fcst_dict[mt]= df_fc
                all_records.append(df_fc)
            else:
                daily_fcst_dict[mt]= pd.DataFrame()

        # show actual
        row_actual= df_sorted[df_sorted["Date"]== day_i]
        if not row_actual.empty:
            print(f"  Actual price on {day_i.date()}: {row_actual['Price'].values[0]:.4f}")

        # 3) daily plots
        if do_daily_plots and daily_fcst_dict:
            plot_daily_two_charts(day_i, horizon_days, daily_fcst_dict, df_sorted)

    final_df= pd.concat(all_records, ignore_index=True) if all_records else pd.DataFrame()
    final_df.sort_values(["Model","BaseDate","ForecastDate"], inplace=True)
    return final_df

# ============================================
# 9. Evaluate & Plot (static approach)
# ============================================
def walk_forward_test(model, df_all_scaled, test_start, window=35):
    """
    Single pass walk-forward, feeding predicted day as next day's input
    (like the original code). 
    """
    df_work= df_all_scaled.copy()
    df_work.sort_values("Date", inplace=True)
    test_dates= np.sort(df_work[df_work["Date"]>= test_start]["Date"].unique())

    model.eval()
    preds=[]
    with torch.no_grad():
        for d in test_dates:
            idx= df_work.index[df_work["Date"]== d]
            if len(idx)==0:
                continue
            idx= idx[0]
            st= idx- window
            if st<0:
                continue
            arr= df_work.iloc[st:idx][features].values
            if np.isnan(arr).any():
                continue
            X_= torch.tensor(arr, dtype=torch.float).unsqueeze(0).to(device)
            p_scaled= model(X_).cpu().numpy()[0]
            df_work.loc[idx, features]= p_scaled
            preds.append((d,*p_scaled))

    dfp= pd.DataFrame(preds, columns=["Date","Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"])
    arr_= dfp[["Pred_Price_scaled","Pred_Open_scaled","Pred_High_scaled","Pred_Low_scaled"]].values
    arr_un= inverse_transform_4cols(arr_)
    dfp["Pred_Price_unscaled"]= arr_un[:,0]
    dfp["Pred_Open_unscaled"]=  arr_un[:,1]
    dfp["Pred_High_unscaled"]=  arr_un[:,2]
    dfp["Pred_Low_unscaled"]=   arr_un[:,3]
    return dfp.sort_values("Date")

def evaluate_and_plot_all_models(
    results_dict, 
    df_scaled, 
    df_train, 
    train_cutoff, 
    val_cutoff, 
    test_cutoff
):
    df_real= df_scaled[["Date","Price","Open","High","Low"]].copy()
    df_real.sort_values("Date", inplace=True)
    mask_trainval= (df_real["Date"]< test_cutoff)
    df_real_plot= df_real[mask_trainval].copy()
    arr_scaled= df_real_plot[features].values
    arr_unscaled= inverse_transform_4cols(arr_scaled)
    df_real_plot["Real_Price_unscaled"]= arr_unscaled[:,0]

    plt.figure(figsize=(14,7))
    plt.plot(df_real_plot["Date"], df_real_plot["Real_Price_unscaled"], color="black", label="Actual (Train+Val)")

    merged_df= df_real.copy()
    merged_df.sort_values("Date", inplace=True)

    color_map= {
        "LSTM": "red",
        "GRU": "blue",
        "RNN": "green",
        "CNN": "orange",
        "EnhancedLSTM": "magenta",
        "Transformer": "cyan",
        "N-BEATS": "brown",
        "N-HITS": "pink",
        "SVM": "olive",
        "GPR": "teal",
        "Boost": "purple"
    }

    for mname, info in results_dict.items():
        val_df= info["val_pred_df"]
        test_df= info["test_pred_df"]
        frames=[]
        if val_df is not None:
            frames.append(val_df.rename(columns={"Pred_Price_unscaled":f"Pred_{mname}"}))
        if test_df is not None:
            frames.append(test_df.rename(columns={"Pred_Price_unscaled":f"Pred_{mname}"}))
        if frames:
            comb= pd.concat(frames, ignore_index=True)
            comb= comb[["Date",f"Pred_{mname}"]].drop_duplicates("Date")
            comb.sort_values("Date", inplace=True)
            merged_df= pd.merge(merged_df, comb,on="Date", how="left")

    for mname in results_dict.keys():
        col= f"Pred_{mname}"
        if col in merged_df.columns:
            plt.plot(
                merged_df["Date"],
                merged_df[col],
                color=color_map.get(mname,"gray"),
                linestyle="-",
                label=f"{mname} Pred"
            )

    # shading
    plt.axvspan(df_train["Date"].min(), train_cutoff, color="skyblue", alpha=0.1, label="Train")
    plt.axvspan(train_cutoff, val_cutoff, color="green", alpha=0.1, label="Validation")
    plt.axvspan(test_cutoff, merged_df["Date"].max(), color="yellow", alpha=0.1, label="Test")

    plt.title("Joint Diagram: Actual Price vs. Predictions (Silver Futures)")
    plt.xlabel("Date")
    plt.ylabel("Price (unscaled)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Zoom for test
    df_test_zoom= merged_df[merged_df["Date"]>= test_cutoff]
    if not df_test_zoom.empty:
        plt.figure(figsize=(12,5))
        for mname in results_dict.keys():
            col= f"Pred_{mname}"
            if col in df_test_zoom.columns:
                plt.plot(df_test_zoom["Date"], df_test_zoom[col], linestyle="--", label=f"{mname} Pred")
        # overlay actual from CSV
        df_full= pd.read_csv(csv_file_complete)
        df_full["Date"]= pd.to_datetime(df_full["Date"], errors="coerce")
        for c in ["Vol.","Change %"]:
            if c in df_full.columns:
                df_full.drop(columns=[c], inplace=True, errors="ignore")
        df_full.sort_values("Date", inplace=True)
        df_test_only= df_full[df_full["Date"]>= test_cutoff].copy()
        plt.plot(df_test_only["Date"], df_test_only["Price"], color="black", lw=2, label="Actual Test Price")

        plt.title("Test Period Zoom-In: Actual vs. Model Predictions")
        plt.xlabel("Date")
        plt.ylabel("Price (unscaled)")
        plt.legend(fontsize=8)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# ============================================
# 10. Final plot of rolling forecasts
# ============================================
def plot_final_rolling_forecasts(final_rolling_fcst, df_actual, 
                                 from_date=pd.to_datetime("2025-01-02"), 
                                 to_date=pd.to_datetime("2025-03-01")):
    if final_rolling_fcst.empty:
        print("No final rolling forecast data. Exiting.")
        return

    df_act= df_actual.copy()
    df_act.sort_values("Date", inplace=True)
    mask= (df_act["Date"]>= from_date)&(df_act["Date"]<= to_date)
    df_aplot= df_act[mask].copy()

    plt.figure(figsize=(12,5))
    plt.plot(df_aplot["Date"], df_aplot["Price"], color="black", label="Actual Price")

    color_map= {
        "LSTM": "red",
        "GRU": "blue",
        "RNN": "green",
        "CNN": "orange",
        "EnhancedLSTM": "magenta",
        "Transformer": "cyan",
        "N-BEATS": "brown",
        "N-HITS": "pink",
        "SVM": "olive",
        "GPR": "teal",
        "Boost": "purple"
    }
    models= final_rolling_fcst["Model"].unique()
    for m in models:
        df_m= final_rolling_fcst[final_rolling_fcst["Model"]== m].copy()
        df_m= df_m[(df_m["ForecastDate"]>= from_date)&(df_m["ForecastDate"]<= to_date)]
        if not df_m.empty:
            plt.plot(
                df_m["ForecastDate"],
                df_m["Pred_Price_unscaled"],
                color=color_map.get(m,"gray"),
                linestyle="--",
                label=f"{m} Forecast"
            )

    plt.title(f"Rolling Forecast from {from_date.date()} to {to_date.date()}")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ============================================
# MAIN
# ============================================
if __name__=="__main__":
    # 1) Prepare
    df_train = df[df["Date"] < pd.to_datetime("2024-08-01")]
    # 2) Model List
    all_model_types = ["LSTM","GRU","RNN","CNN","Transformer","N-BEATS","N-HITS","SVM","Boost"]
    results_dict={}

    # 3) Train or Load (static)
    print("===== START TRAINING OR LOADING MODELS (STATIC CUTOFF) =====")
    for mt in all_model_types:
        print(f"\n=== Checking {mt} model with original cutoffs ===")

        out_filename = os.path.join(best_models_dir, f"best_{mt}_Silver_V1")
        if mt in ["SVM","Boost","GPR"]:
            out_filename += ".pkl"
            model = train_or_load_sklearn(
                model_type=mt,
                df_scaled_in=df_scaled,
                train_cutoff_in=pd.to_datetime("2024-08-01"),
                val_cutoff_in=pd.to_datetime("2024-12-31"),
                model_save_name=out_filename,
                window_size=WINDOW_SIZE
            )
        else:
            out_filename += ".pt"
            model = train_or_load_torch(
                model_type=mt,
                window_size=WINDOW_SIZE,
                df_scaled_in=df_scaled,
                train_cutoff_in=pd.to_datetime("2024-08-01"),
                val_cutoff_in=pd.to_datetime("2024-12-31"),
                model_save_name=out_filename,
                num_epochs=NUM_EPOCHS
            )

        # Validation predictions
        if mt in ["SVM","Transformer","Boost","GPR"]:
            val_df= iterative_half_blind_validation_preds(
                model, df_scaled, pd.to_datetime("2024-08-01"), pd.to_datetime("2024-12-31"), WINDOW_SIZE
            )
        else:
            val_df= half_blind_validation_preds_df(
                model, df_scaled, pd.to_datetime("2024-08-01"), pd.to_datetime("2024-12-31"), WINDOW_SIZE
            )

        # Test predictions
        test_df= walk_forward_test(
            model, df_scaled, pd.to_datetime("2025-01-01"), WINDOW_SIZE
        )

        results_dict[mt] = {
            "val_pred_df": val_df,
            "test_pred_df": test_df
        }

    # 4) Evaluate & Plot (static approach)
    evaluate_and_plot_all_models(
        results_dict, 
        df_scaled, 
        df_train, 
        train_cutoff = pd.to_datetime("2024-08-01"), 
        val_cutoff   = pd.to_datetime("2024-12-31"), 
        test_cutoff  = pd.to_datetime("2025-01-01")
    )

    # 5) Rolling Approach
    df_complete_all= pd.read_csv(csv_file_complete)
    df_complete_all["Date"]= pd.to_datetime(df_complete_all["Date"], errors="coerce")
    for cdrop in ["Vol.","Change %"]:
        if cdrop in df_complete_all.columns:
            df_complete_all.drop(columns=[cdrop], inplace=True, errors="ignore")
    for c2 in features:
        df_complete_all[c2]= df_complete_all[c2].astype(str).str.replace(",", "", regex=True).astype(float)
    df_complete_all.sort_values("Date", inplace=True)
    df_complete_all.reset_index(drop=True, inplace=True)

    print("\n===== START ROLLING TRAIN/VALIDATION/PREDICTION from 2025-01-02 to 2025-01-31 =====")
    final_rolling_fcst= rolling_train_validate_predict_moe(
        df_full= df_complete_all,
        scaler_obj= scaler,
        model_types= all_model_types,
        start_train_cutoff= pd.to_datetime("2024-08-01"),
        start_val_cutoff=   pd.to_datetime("2024-12-31"),
        start_pred=         pd.to_datetime("2025-01-02"),
        end_pred=           pd.to_datetime("2025-01-31"),
        horizon_days= 30,
        do_daily_plots=True
    )
    print("\n===== ROLLING FORECAST OUTPUT =====")
    print(final_rolling_fcst.head(50))
    print("Total rows in final_rolling_fcst:", len(final_rolling_fcst))

    # 6) Final "concatenated" diagrams
    print("\n===== FINAL CONCATENATED DIAGRAMS =====")
    plot_final_rolling_forecasts(
        final_rolling_fcst,
        df_complete_all,
        from_date=pd.to_datetime("2025-01-02"),
        to_date=pd.to_datetime("2025-03-01")
    )
    plot_final_rolling_forecasts(
        final_rolling_fcst,
        df_complete_all,
        from_date=pd.to_datetime("2025-02-01"),
        to_date=pd.to_datetime("2025-03-01")
    )

    print("\nAll done.")
```
