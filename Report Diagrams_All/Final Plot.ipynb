```python
import matplotlib.pyplot as plt, numpy as np, matplotlib as mpl

# matrix of means
means = np.array([[-0.72, -0.72, 0.37],
                  [0.60,  1.63, 3.66],
                  [2.88,  2.88, 2.49],
                  [0.43,  2.16, 2.93]])

row_labels = ['ARIMA Baseline Model',
              'Selected Model Combinations',
              'MoE of Selected Models',
              'Hybrid of MoE and Selected Models']

col_labels = ['Naive Conservative',
              'Naive Optimistic',
              'Dynamic Confidence Voting']

# reverse rows for display
M   = means[::-1]
r_l = row_labels[::-1]

fig, ax = plt.subplots(figsize=(6,4))
im  = ax.imshow(M, cmap='coolwarm')

# annotate every non-merged cell
for i in range(4):
    for j in range(3):
        if (r_l[i] in ['ARIMA Baseline Model','MoE of Selected Models']) and j in (0,1):
            continue  # will annotate once in centre
        ax.text(j, i, f"{M[i,j]:.2f}", ha='center', va='center', color='black', fontsize=8)

# merged blocks
merges = {'ARIMA Baseline Model': -0.72,
          'MoE of Selected Models': 2.88}

for label, val in merges.items():
    i = r_l.index(label)
    # draw outline
    ax.add_patch(plt.Rectangle((-0.5,i-0.5), 2, 1,
                 fill=False, edgecolor='black', linewidth=1))
    # central annotation
    ax.text(0.5, i, f"{val:.2f}", ha='center', va='center',
            color='black', fontsize=8)

ax.set_xticks(range(3))
ax.set_xticklabels(col_labels, rotation=20, ha='right')
ax.set_yticks(range(4))
ax.set_yticklabels(r_l)
plt.colorbar(im, label="Mean Monthly P&L Return (%)")
ax.set_title("Mean Monthly P&L Return (%)\nPricePrediction Philosophy  Trading Strategy")
plt.tight_layout()
plt.savefig('final_heatmap.png', dpi=500)
plt.show()
```
